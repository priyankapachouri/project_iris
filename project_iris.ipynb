```python

file_path = "C:\\Users\\aksac\\Downloads\\iris1.csv"

try:
    df = pd.read_csv(file_path)

    if 'Sepal Length' in df.columns:
        sepal_length_sum = df['Sepal Length'].sum()
        print("Sum of 'Sepal Length' column:", sepal_length_sum)
        
        sepal_length_mean = df['Sepal Length'].mean()
        print("Mean of 'Sepal Length' column:", sepal_length_mean)
        sepal_length_mode = df['Sepal Length'].mode().values[0]
        print("Mode of 'Sepal Length' column:", sepal_length_mode)
    else:
        print("'Sepal Length' column not found in the DataFrame.")
        
except FileNotFoundError:
    print(f"File '{file_path}' not found.")
except Exception as e:
    print(f"Error occurred: {str(e)}")
```

```python

df = pd.read_csv("C:\\Users\\aksac\\Downloads\\iris1.csv")

print("Missing values before cleaning:")
print(df.isnull().sum())

df['Sepal Length'].fillna(df['Sepal Length'].mean(), inplace=True)
df['Sepal Width'].fillna(df['Sepal Width'].mean(), inplace=True)
df['Petal Length'].fillna(df['Petal Length'].mean(), inplace=True)
df['Petal Width'].fillna(df['Petal Width'].mean(), inplace=True)

print("\nMissing values after cleaning:")
print(df.isnull().sum())

df.to_csv("C:\\Users\\aksac\\Downloads\\iris1_cleaned.csv", index=False)
```

```python
numerical_cols = df.select_dtypes(include=[np.number]).columns
df_numerical = df[numerical_cols]

corr_matrix = df_numerical.corr()
plt.figure(figsize=(5,5))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)

plt.show()
```
